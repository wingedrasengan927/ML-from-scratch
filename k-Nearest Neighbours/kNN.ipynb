# Preparing the Data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# let's load the text file which contains the dataset and parse it
# the last movie type in the dataset is unknown and we'll classify that
import re

with open('movie dataset.txt', 'r') as f:
    dataset = []
    for line in f.readlines():
        # let's separate numbers from strings
        parse_line = re.split('([0-9]+)', line)
        # let's remove the empty strings
        parse_line = [x for x in parse_line if x != " "]
        # let's strip the whitespace; strip() also removes the character '\n'
        parse_line = [x.strip() for x in parse_line]
        # and now, let's append it to a dataset
        dataset.append(parse_line)
dataset
```

```python
# let's make a dataframe of the data
movie_dataset = pd.DataFrame(dataset, columns=['Movie', 'Number of kicks',
                                              'Number of Kisses', 'Type of Movie'])
movie_dataset
```

Imagine each observation to be a data point in the space, with the co-ordinates as it's features. kNN works by measuring the distance from the point we want to classify to all the other points in the space and then based on the sorted distance, it finds the k nearest neighbours and returns the majority class among these neighbours.

The euclidean distance between the points $(A_{0},\ A_{1}),\ (B_{0},\ B_{1})$ is given by:
<br>$\sqrt{(A_{0}-B_{0})^{2}+(A_{1}-B_{1})^{2}}$


```python
# the numbers present here are of type str.let's convert them into integers
movie_dataset.iloc[:, [1, 2]] = movie_dataset.iloc[:, [1, 2]].applymap(lambda x: int(x))
```

```python
# let's look at the features we required
movie_dataset.iloc[:, [1, 2]]
```

```python
# let's define a function to compute the euclidean distance

# it takes two arguments:
# the datapoint we want to classify and the datapoint to which we want to find the distance 
# it outputs the euclidean distance 
def euclidean_distance(classify_point, other_point):
    return np.sqrt(sum((classify_point-other_point)**2))
```

```python
# let's test the function
a = np.array([0, 0])
b = np.array([1, 1])
euclidean_distance(a, b)
```

```python
# let's use the apply method to find distances from each of the datapoints
distance = movie_dataset.iloc[:-1, [1, 2]].apply(lambda x: euclidean_distance(movie_dataset.iloc[-1, [1, 2]], x),
                                                 axis=1)
distance
```

```python
# now let's insert the distance column in the second last position
movie_dataset_train = movie_dataset.iloc[:-1, :].copy()
# Note: say if we are inserting at location 1, then the column already present there shifts to the right
movie_dataset_train.insert(movie_dataset.shape[1]-1, "distance", distance)
movie_dataset_train
```

```python
# let's sort the class labels column by the distances
movie_dataset_train.sort_values(by=movie_dataset_train.columns[-2])
```

```python
# Now from this sorted dataset, let's pick the k(say 3) nearest neighbours and return the majority class
movie_dataset_train.sort_values(by=movie_dataset_train.columns[-2]).iloc[:3, -1].value_counts(sort=True, 
                                                                                            ascending=False).index[0]
```

```python
# let's drop the distance column as we don't require it anymore
movie_dataset_train.drop("distance", axis=1, inplace=True)
```

```python
movie_dataset_train
```

# Putting the kNN Classification Algorithm into Action

```python
# EDIT : The version 2 which uses arrays as an argument is efficent and simple. It's given below in the end
# however, use this if data is presented in the form of a dataframe

# now let's put it all together

# one assumption is that the dataset is a pandas dataframe and the point to classify is a series
# and the class labels is the last column

# it takes 4 arguments:
# 1. the point we want to classify
# 2. the training dataset
# 3. the index of what features to be used 
# 4. the number of neighbours - k
def classifykNN(to_classify, train_dataset, feature_index_list, k):
    # let's make a copy of the dataframe so that the original doesn't change
    train_dataset_copy = train_dataset.copy()
    # let's calculate the euclidean distance
    distance = train_dataset.iloc[:, feature_index_list].apply(lambda x: euclidean_distance(to_classify, x), axis=1)
    # let's append this distance array to the dataframe at the second last position
    train_dataset_copy.insert(train_dataset_copy.shape[1]-1, "Distance", distance)
    # let's sort the dataframe by the distance
    train_dataset_copy = train_dataset_copy.sort_values(by=train_dataset_copy.columns[-2])
    # let's return the majority class among k nearest neighbours
    return train_dataset_copy.iloc[:k, -1].value_counts(sort=True, ascending=False).index[0]
    
```

```python
# let's test our kNN Classifier
classifykNN(movie_dataset.iloc[-1, :], movie_dataset_train, [1, 2], 3)
```

# Normalizing the data

the euclidean distance between person 2 and 3 would be $\sqrt{(26052-75136)^{2} + (1.44-13.14)^{2} + (0.80-0.42)^{2}}$
<br>Now if you observe, the factor $(26052-75136)^{2}$ is very large and is going to make the most difference, and consequently will dominate over others.
<br>Just because it's values are large, it shouldn't have any extra importance. Hence the need for normalization.

When dealing with values that lie in different ranges, it’s common to normalize them.
Common ranges to normalize them to are 0 to 1 or -1 to 1. To scale everything from 0
to 1, you need to apply the following formula:<br>

$newValue\ =\ \dfrac{(oldValue\ -\ min)}{(max\ -\ min)}$


```python
# normalization in action
# let's see with an example

movie_dataset_train
```

```python
# let's just check if the apply method works for normalization
movie_dataset_train.iloc[:, [1, 2]].apply(lambda x: x-x.min()) # it works
```

```python
# let's use the apply method to normalize
dating.iloc[:, :-1] =  dating.iloc[:, :-1].apply(lambda x: (x-x.min())/(x.max()-x.min()))
dating.head()
```

# Prepare: converting images into test vectors

```python
# let's see if the shape of vector matters in distance calculation
# we see that both are the same. Hence the shape of an array  doesn't matter in distance calculation

a = np.array([[1, 2], [3, 4]]) - np.array([[1, 1], [1, 1]])
print((a**2).sum())
b = np.array([1, 2, 3, 4]) - np.array([1, 1, 1, 1])
print((b**2).sum())
```

```python
# let's convert this into a 32*32 matrix
def img2vector(f): # takes the file object as an argument
    returnVec = np.zeros((32, 32))
    for i in range(32): # iterating through all the rows
        line = f.readline()
        for j in range(32): # iterating through all the columns
            returnVec[i, j] = int(line[j])
    return returnVec
```

```python
# let's test our function 
with open('digits/trainingDigits/0_0.txt', 'r') as f:
    print(img2vector(f))
```

```python
# let's see the list of documnets in the directory
import os
for doc in os.listdir('digits/trainingDigits'):
    print(doc)
```

```python
# we need to define euclidean distance v2 for our new kNN classifier
def euclidean_distance_v2(classify_point, other_point): # both arguments should be numpy arrays of same shape
    return np.sqrt(np.square(classify_point-other_point).sum())
```

```python
# let's test our function
euclidean_distance_v2(np.array([[1, 2], [3, 4]]), np.array([[2, 4], [3, 7]]))
```

```python
# since it'll be difficult to use a dataframe dependent kNN Classifier for this classifier,
# let's instead modify the classifier to accomodate arrays as arguments

from collections import Counter

def classifyKNN_v2(classify_point, dataset, labels_list, k):
    # first we calculate distances and append it to an array
    distance = np.zeros(len(dataset))
    for i in range(len(dataset)):
        distance[i] = euclidean_distance_v2(classify_point, dataset[i])
    # now let's sort this distance array and return the indices of the corresponding distances
    sortedDistanceIndices = np.argsort(distance)
    # now let's take the labels of top k distances and return the majority class
    topLabelsList = labels_list[sortedDistanceIndices[:k]]
    return Counter(topLabelsList).most_common()[0][0]
    
```

# Test: kNN on handwritten digits

```python
# let's prepare the training data set

import os

list_of_docs = os.listdir('digits/trainingDigits')
training_data = []
for i in range(len(list_of_docs)):
    with open('digits/trainingDigits/{}'.format(list_of_docs[i]), 'r') as f:
        training_data.append(img2vector(f))
training_data = np.array(training_data)
```

```python
# let's prepare the corresponding labels list

labels_list = []
for doc in list_of_docs:
    labels_list.append(doc.split('_')[0])
labels_list = np.array(labels_list)
labels_list
```

```python
# Now let's test our classifier and find the error rate

error_rate = 0
list_of_testDocs = os.listdir('digits/testDigits')
for i in range(len(list_of_testDocs)):
    with open('digits/testDigits/{}'.format(list_of_testDocs[i]), 'r') as f:
        classify_label = classifyKNN_v2(img2vector(f), training_data, labels_list, 3)
        actual_label = list_of_testDocs[i].split('_')[0]
        if actual_label != classify_label:
            print("the actual digit is {}. the classified digit is {}".format(actual_label, classify_label))
            error_rate += 1
print()
print("the error rate is {}".format(error_rate/len(list_of_testDocs)))
```

kNN is
an example of instance-based learning, where you need to have instances of data close
at hand to perform the machine learning algorithm. The algorithm has to carry
around the full dataset; for large datasets, this implies a large amount of storage. In
addition, you need to calculate the distance measurement for every piece of data in
the database, and this can be cumbersome.
