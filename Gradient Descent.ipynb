```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

The gradient descent is given by
![gradient descent](GD.png)

Let's take an arbitary cost function and find it's minimum value using gradient descent.
<br>cost function: $x^{3} -3x^{2} + 7$
<br>Note that here there is only one parameter to optimize.
<br>The derivative of this function would be $3x^{2} - 6x$

```python
def cost_func(x):
    '''The cost function. Returns output value for a given input'''
    return math.pow(x, 3) - 3*math.pow(x, 2) + 7
```

```python
# let's plot the function in range [-1, 3]
x = np.linspace(-1, 3, 1000)
y = np.array(list(map(cost_func, x)))
plt.figure(figsize=(6, 4))
plt.plot(x, y)
```

```python
def deriv(x):
    '''this function returns the ouput value of derivative of the cost function for a particular input'''
    return 3*(np.power(x, 2)) - 6*x
```

```python
def step(x_current, alpha, precision=0.001, x_prev=None):
    '''the gradient descent function.
    arguments: 
    x_current - The current value of x
    x_prev - The previous value of x before updating
    alpha - learning rate
    '''
    x_list, y_list = [x_current], [cost_func(x_current)]
    
    # put x prev to any arbitary value and make sure the difference is not zero
    if x_prev is None:
        x_prev = x_current + 1
        
    # keep on updating until we reach the local minima
    limit = 0
    while abs(x_current - x_prev) > precision and cost_func(x_current) > limit:
        x_prev = x_current
        x_current = x_prev - (alpha * deriv(x_prev))
        x_list.append(x_current)
        y_list.append(cost_func(x_current))
        
    plt.figure(figsize=(6, 4))
    x = np.linspace(-1, 3, 1000)
    y = np.array(list(map(cost_func, x)))
    plt.plot(x, y)
    plt.scatter(x_list, y_list, c='g')
    plt.plot(x_list, y_list)
    
    return x_current
```

```python
step(0.5, 0.05)
```
